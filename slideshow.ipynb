## Investigation Overview

We will investigate if employment status has relationship with the rate of charged off / defaulted loan. We will also investigate other variables such as home ownership and prosper rating to see if those factors will affect the borrower rate.

## Dataset Overview

This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from pandas.api.types import CategoricalDtype

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("prosperLoanData.csv")
pd.set_option('display.max_columns', None)

#Change the data type to be categorical. it will then be listed in the FacetGrid in order
cats_to_order = ['AA', 'A', 'B', 'C', 'D', 'E']
covered_type = CategoricalDtype(categories=cats_to_order, ordered=True)
df['ProsperRating (Alpha)'] = df['ProsperRating (Alpha)'].astype(covered_type)
```

## Not Employed and Retired group are having higher defaulted and charged off loan

Comparing not employed and retired group with other group employed and full-time (as well as part-time, self-employed group etc. Not shown here for simplicity), Not Employed group have got the highest Charged off rate and the Retired group comes the next highest. And Retired group have got the highest default rate.

```python
df_by_employ_loan = df.groupby("EmploymentStatus").LoanStatus.value_counts(normalize=True)
df_by_employ_loan = df_by_employ_loan.reset_index(name='normalisedValue')
df_by_employ_loan = df_by_employ_loan[df_by_employ_loan.EmploymentStatus.isin(["Employed", "Full-time", "Not employed", "Retired"])]
df_by_employ_loan = df_by_employ_loan.sort_values(by=['EmploymentStatus', 'LoanStatus'])
g = sb.FacetGrid(data=df_by_employ_loan, col='EmploymentStatus', height=4, col_wrap=2)
g.map(plt.barh, 'LoanStatus', 'normalisedValue', color=sb.color_palette()[8])
g.set_titles('{col_name}', size=15)
for e in [2,3]:
    g.axes.flat[e].set_xlabel("Normalised Value", size=15)
for e in range(0,4):
    g.axes.flat[e].tick_params(axis='both', labelsize=12)
for e in [0,2]:
    g.axes.flat[e].set_ylabel("Loan Status", size=15)
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Loan Status distribution by Employment Status", size=20);
```

## Home Owner receive better Borrower rate

If a customer owns a house, they would receive a lower borrower rate (at around 0.183) than those who do not (around 0.203).

```python
def trans_homeown(x):
    if x == True:
        return 'Home owner'
    else:
        return 'Not a Home owner'
```

```python
plt.figure(figsize=(10,8))
g = sb.barplot(data=df, y=df.IsBorrowerHomeowner.apply(trans_homeown), x='BorrowerRate', ci=None, color=sb.color_palette()[8])
plt.title("Borrower Rate by Home Ownership Status", size=20)
plt.ylabel("")
g.tick_params(axis='both', labelsize=15)
plt.xlabel("Borrower Rate", size=15);
```

## Prosper Rating has critical influence on Borrower Rate

The prosper rating affects the borrower rate. With higher rating, the borrower rate is lower. However, within each rating, we can observe that monthly income does not affect the borrower rate.

```python
g = sb.FacetGrid(data=df.sample(500), col='ProsperRating (Alpha)', col_wrap=3, height=4)
g.map(plt.scatter, 'StatedMonthlyIncome', 'BorrowerRate', color=sb.color_palette()[8])
g.set_titles('{col_name}', size=15)
for e in range(3,6):
    g.axes.flat[e].set_xlabel("Stated Monthly Income", fontsize=15)
for e in [0,3]:
    g.axes.flat[e].set_ylabel("Borrower Rate", fontsize=15)
for e in [0,3,4,5]:
    g.axes.flat[e].tick_params(axis='both', labelsize=10)
plt.subplots_adjust(top=0.85)
plt.suptitle("Borrow Rate vs Monthly Income by Prosper Rating", fontsize=20);
```
