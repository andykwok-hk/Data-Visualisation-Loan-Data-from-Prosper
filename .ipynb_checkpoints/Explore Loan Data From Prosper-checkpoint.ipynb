This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.

```python
df = pd.read_csv("prosperLoanData.csv")
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

In the dataset, we have huge amount of variables and we will focus on a few on them to grasp an overview of the whole dataset. Those variables are Loan Status, Occupation, Employment Status, isborrowerhomeowner and more. We will see how those variables interact with each other through visualising the data.

## Univariate Exploration

```python
plt.figure(figsize=(10,10))
sb.countplot(data=df, y='Occupation', order=df.Occupation.value_counts().index)
```

We have quite a diversed professional backgound for the customer. Top 10 are white collar customers but there are some blue collar customers as well.

```python
sb.countplot(data=df, y='EmploymentStatus', order=df.EmploymentStatus.value_counts().index)
```

Most of the people are employed in different forms (self-employed, full-time, part-time etc.) Just a little proportion of them are not employed and retired.

```python
plt.hist(data=df, x='BorrowerRate', bins=20)
```

Some of them are having higher interest rate then others. We can filter those out and dig deeper in those case to find out the possible relationship with other variables. We will look into those deeper to see if there's any pattern in it like how the loan status distributed when other factors like home ownership, monthly income and employment status kick in.

```python
sb.countplot(data=df, y='LoanStatus', order=df.LoanStatus.value_counts().index)
```

Most of the Loan are current and quite a significant amount are completed. While we can see there is quite a number of charged off and defaulted loan. We will look into those deeper to see if there's any pattern in it like how the loan status distributed when other factors like home ownership, monthly income and employment status kick in.

```python
bin_edges = np.arange(0,40000, 1000)
```

```python
plt.hist(data=df, x='StatedMonthlyIncome', bins=bin_edges)
```

Bins edges are used to zoom in the lower amount of monthly income which represent most of the data. We can see that customers mostly having around 5000 monthly income. We can later look into whether it have relationship with borrower rate. And loan origin amount.

## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your
data. Make sure the variables that you cover here have been introduced in some
fashion in the previous section (univariate exploration).

```python
df_employ_sub = df[df.LoanStatus.isin(["Defaulted","Chargedoff"])]
```

```python
plt.figure(figsize=(10,10))
sb.countplot(data=df_employ_sub, y='LoanStatus', hue='EmploymentStatus')
```

For those defaulted and charged off loans, most of the people are employed. But it may well be due to large amount of customers are employed. So proportion representation may be more informative.

```python
df_by_employ_loan = df.groupby("EmploymentStatus").LoanStatus.value_counts(normalize=True)
```

```python
df_by_employ_loan = df_by_employ_loan.reset_index(name='normalisedValue')
df_by_employ_loan = df_by_employ_loan.sort_values(by=['EmploymentStatus', 'LoanStatus'])
```

```python
df_by_employ_loan.tail(10)
```

```python
g = sb.FacetGrid(data=df_by_employ_loan, col='EmploymentStatus', col_wrap=3)
g.map(plt.barh, 'LoanStatus', 'normalisedValue')
```

The defaulted proportion is the highest in retired group and the charged off proportion is the highest in unemployed group.

```python
plt.scatter(data=df.sample(500), x='StatedMonthlyIncome', y='BorrowerRate', alpha=0.7)
```

Customers do not get better borrower rate if they have a higher income. borrower rate are cluttered from 0.05 to 0.35. Maybe it is more related to their credit score. We will add it to be our third variable to investigate more about the relationship.

```python
def trans_homeown(x):
    if x == True:
        return 'Home owner'
    else:
        return 'Not a Home owner'
```

```python
sb.barplot(data=df, y=df.IsBorrowerHomeowner.apply(trans_homeown), x='BorrowerRate')
```

If a customer owns a house, they would receive a lower borrower rate (at around 0.183) than those who do not (around 0.203).

## Multivariate Exploration

```python
#Change the data type to be categorical. it will then be listed in the FacetGrid in order
from pandas.api.types import CategoricalDtype
cats_to_order = ['AA', 'A', 'B', 'C', 'D', 'E']
covered_type = CategoricalDtype(categories=cats_to_order, ordered=True)
```

```python
df['ProsperRating (Alpha)'] = df['ProsperRating (Alpha)'].astype(covered_type)
```

```python
g = sb.FacetGrid(data=df.sample(500), col='ProsperRating (Alpha)', col_wrap=3)
g.map(plt.scatter, 'StatedMonthlyIncome', 'BorrowerRate')
```

Monthly income does not affect the borrower rate. However, the prosper rating do affect the borrower rate. With higher rating, the borrower rate is lower.
